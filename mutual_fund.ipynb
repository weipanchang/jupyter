```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
# for a in os.listdir(".\\Stock\\"):
#     print (a )
```

```python
# Function to read csv file
def read_file():

    file_path =  ".\\Stock\\"
    df1 = pd.DataFrame()    
    for f in os.listdir(file_path):
            print (f)
            file_open = open(file_path + f,"r")
            column_list = []
            return_list = []
            for line in iter(file_open):
                line1 = line.rstrip()
                line2 = line1.replace("Large-Cap Growth","0")
                line3 = line2.replace("Large-Cap Blend","1")
                line4 = line3.replace("Large-Cap Value","2")
                line_split=line4.split(",")
                column_list.append(line_split[0])
                if line_split[1] != 'N/A':
#                    print(type(float(line_split[1][:-1])))
                    return_list.append(line_split[1][:-1])
                else:
                    return_list.append(line_split[1])
            column_list.append('group')
            return_list.append(line_split[2])
            df = pd.DataFrame([i  for i in return_list]).T
            df.columns = pd.DataFrame(column_list)
            df1 = pd.concat([df1, df], axis=0, join='outer', ignore_index=True)
    return df1
```

```python
# Read the training dataset

df = read_file()

# Drop N/A column
df.dropna(axis='columns', how='any', inplace=True)

# Drop year column
df.drop(['Year'], axis = 1, inplace = True)
df

#print (df.iloc[:10])
```

```python
# Convert data to numpy array
dataset = np.array(df)
X2 = dataset[:,0:10]
y2 = dataset[:,10]
```

```python
# Large-Cap Growth -- 0
# Large-Cap Blend -- 1
# Large-Cap Value -- 2
features = X
target = y

f, ax = plt.subplots(1)
for i in np.unique(target):
    mask = (target == i)
    print(features[mask,0])
    plt.scatter(features[mask, 0], features[mask, 8], label=i)
ax.legend()
plt.xlabel('2009')
plt.ylabel('2017')
plt.show()
```

```python
# Logistic Regression
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
# print(model.intercept_)
# print(model.coef_)
```

```python
def plot_cm(cm):   
    groups = ["Large-Cap Growth","Large-Cap Blend","Large-Cap Value"]
    df_cm = pd.DataFrame(cm, index=groups, columns=groups)
    fig = plt.figure(figsize=(5,5))
    heatmap = sns.heatmap(df_cm, annot=True, fmt="d",cmap="YlGnBu",linecolor="blue",linewidths=1.0)
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=10)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=10)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
plot_cm(cm)
```

```python
## Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
plot_cm(cm)
```

```python
# Knn- classification
# k-Nearest Neighbor Classification
from sklearn.neighbors import KNeighborsClassifier
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```

```python
plot_cm(cm)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
print(model)
# make predictions
expected = y
predicted = model.predict(X)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
cm = metrics.confusion_matrix(expected, predicted)
```
